Create a BookStack class to manage the stack operations

```python
class bookstack:
    def __init__(self):
        self.stack = []
    def push(self,book):
        self.stack.append(book)
    def pop(self):
       if not self.is_empty():
        return self.stack.pop()
       else:
           return "stack is empty"
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        else:
            return "stack is empty"
    def is_empty(self):
        return len(self.stack) == 0
    def size(self):
        return len(self.stack)
```

Create a subclass EBook that inherits from the Book class. Add an additional attribute file_size to represent the size of the eBook file. Override the __str__ method to include the file size in the string representation

```python
class Ebook(book): 
    def __init__(self, title, author, genre, publisher, price, ebook_format, file_size):
        super().__init__(title, author, genre)
        self.file_size = file_size
        self.ebook_format = ebook_format
        self.price = price
        self.publisher = publisher
   
    def __str__(self):
        return f"{super().__str__()}, Price: ${self.price}, File Size: {self.file_size} MB"
        
my_ebook = Ebook("The Great Gatsby", "F. Scott Fitzgerald", "fiction", "Penguin", 12.99, "PDF", 57.79)
print(my_ebook.title)  
print(my_ebook.ebook_format)  
print(my_ebook)
```

Create a Library class that contains a list of BookStack objects.Implement methods to add a new stack, remove a stack, and list all stacks in the library

```python
class bookstack:
    def __init__(self, name):
        self.name = name
        self.books = []
    def add_book(self, book):
        self.books.append(book)
    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
        else:
            print(f"{book} not found in self.name")
    def list_books(self):
        print(f"books in {self.name}:")
        for book in self.books:
            print(f"-{book}")

class library:
    def __init__(self):
        self.stacks = {}
    def add_stack(self, stack_name):
        if stack_name not in self.stacks:
            self.stacks[stack_name] = bookstack(stack_name)
        else:
            print(f"{stack_name} already exists")
    def remove_stack(self,stack_name):
        if stack_name in self.stacks:
            del self.stacks[stack_name]
        else:
            print(f"{stack_name} not found")
    def list_stacks(self):
        print("Available stacks:")
        for stack_name in self.stacks:
            print(f"- {stack_name}")

my_library = library()
my_library.add_stack("stack1")
my_library.add_stack("stack2")

stack1_stack = my_library.stacks["stack1"]
stack1_stack.add_book("The Great Learning")
stack1_stack.add_book("The Doctrine of the Mean")

stack2_stack = my_library.stacks["stack2"]
stack2_stack.add_book("Confucian Analects")
stack2_stack.add_book("The Works of Mencius")

my_library.list_stacks()
stack1_stack.list_books()
stack2_stack.list_books()
```

Modify the BookStack class to raise a custom exception StackEmptyError when attempting to pop or peek from an empty stack. Handle this exception gracefully in your main program

```python
class StackEmptyError(Exception):
    pass

class BookStack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.items:
            raise StackEmptyError("Cannot pop from an empty stack")
        return self.items.pop()

    def peek(self):
        if not self.items:
            raise StackEmptyError("Cannot peek an empty stack")
        return self.items[-1]

try:
    stack = BookStack()
    stack.push("Book 1")
    stack.push("Book 2")
    stack.pop()  
    stack.pop() 
    stack.pop() 
except StackEmptyError as e:
    print(f"Error: {e}")
```

Add a class method from_string to the Book class that creates a Book instance from a string in the format “title,author”. Add a static method is_valid_title that checks if a given title is valid (e.g., notempty)

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    @classmethod
    def from_string(cls, book_string):
        title, author = book_string.split(",")
        return cls(title.strip(), author.strip())

    @staticmethod
    def is_valid_title(title):
        return bool(title.strip())  

book_string = "You can do this, Karen Hodges Miller"
book_instance = Book.from_string(book_string)
print(f"Title: {book_instance.title}, Author: {book_instance.author}")

title_to_check = "   " 
print(f"Is title valid? {Book.is_valid_title(title_to_check)}")
```
